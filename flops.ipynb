# facebook/MobileLLM-125M

```python
#Large Languase Model, such as llama2-7b.
from calflops import calculate_flops
from transformers import LlamaTokenizer
from transformers import LlamaForCausalLM

batch_size, max_seq_length = 1, 128
model_name = "facebook/MobileLLM-125M"

model = LlamaForCausalLM.from_pretrained(model_name)
tokenizer = LlamaTokenizer.from_pretrained(model_name)
flops, macs, params = calculate_flops(model=model,
                                      input_shape=(batch_size, max_seq_length),
                                      transformer_tokenizer=tokenizer)
print("Mobilellm FLOPs:%s   MACs:%s   Params:%s \n" %(flops, macs, params))
#Llama2(7B) FLOPs:1.7 TFLOPS   MACs:850.00 GMACs   Params:6.74 B
```

```python
import torch
from torch.profiler import profile, record_function, ProfilerActivity
from transformers import AutoModel


batch_size = 1
max_seq_length = 128
vocab_size = model.vocab_size  # for BERT tokenizer

# Simulate token IDs (int64)
inputs = torch.randint(low=0, high=vocab_size, size=(batch_size, max_seq_length))

with profile(activities=[ProfilerActivity.CPU], record_shapes=True) as prof:
    with torch.no_grad():
        outputs = model(inputs)

print(prof.key_averages().table(sort_by="cpu_time_total", row_limit=10))
```
